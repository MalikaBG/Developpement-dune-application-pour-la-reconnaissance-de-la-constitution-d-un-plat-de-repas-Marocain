```python
def Pomme_De_Terre_Classifier(path):
    img = cv2.imread(path)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    out = cv2.resize(imread(path),(400,600))
    src.new_frame(out)
    if classifier.predict(src.slice(0, 0, 400, 600)) == 1 :
        print("- Pomme de Terre")
```
